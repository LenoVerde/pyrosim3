```python
import sys
import notebookutil as nbu
sys.meta_path.append(nbu.NotebookFinder())
from quadrupedIndivisual import INDIVIDUAL

import copy
import random

class POPULATION:
    
    def __init__(self, popSize):
        self.p = {}
        self.popSize= popSize
        '''


    def __init__(self):
        self.p = {}
        popSize = 5
        for i in range(0,popSize):
              self.p[i] = INDIVIDUAL(i)
    ''' 
        
    def Print(self):
        for i in self.p:
            print('[{0} {1}]'.format(self.p[i].ID, self.p[i].fitness),end="")
        print("")
            
    def Evaluate(self):#ビデオオフ
        for i in self.p:
            self.p[i].Evaluate(True)
        for i in self.p:
            self.p[i].Compute_Fitness()
            
    def Evaluate3(self):#ビデオオフ
        
        self.sim = pyrosim.Simulator( play_paused=False , eval_time=2000 ,play_blind = False )
        for i in self.p:
            self.p[i].robot = ROBOT(self.sim, self.genome )

        self.sim.start()
        
        for i in self.p:
            self.p[i].Compute_Fitness()
            
    def Evaluate2(self):#ビデオオン
        for i in self.p:
            self.p[i].Evaluate(False)
        for i in self.p:
            self.p[i].Compute_Fitness()
            
    def Mutate(self):
            for i in self.p:
                self.p[i].Mutate()
                
    def ReplaceWith(self,other):
        for i in self.p:
            if ( self.p[i].fitness < other.p[i].fitness ):
                self.p[i] = other.p[i]

    def Initialize(self):
        for i in range(0, self.popSize):
            self.p[i]= INDIVIDUAL(i)
                
    def Fill_From(self , other):
        self.Copy_Best_From(other)
        self.Print()
        self.Collect_Children_From(other)
        self.Print()
    
    def Copy_Best_From(self, other):
        highest_fitness=-1
        highest_id= -1
        for i in other.p:
            if (other.p[i].fitness>highest_fitness):
                highest_fitness=other.p[i].fitness
                highest_id=i
        best= copy.deepcopy(other.p[highest_id])
        self.p[0]= best

    def Collect_Children_From(self, other):
        for i in range(1, self.popSize, 1):
            winner = other.Winner_Of_Tournament_Selection()
            self.p[i]= copy.deepcopy(winner)
            self.p[i].Mutate()

    def Winner_Of_Tournament_Selection(self):
        p1= random.randint(0, self.popSize-1)
        p2= random.randint(0, self.popSize-1)
        while (p1==p2):
            p2= random.randint(0, self.popSize-1)
        if(self.p[p1].fitness>self.p[p2].fitness):
            return(self.p[p1])
        else:
            return(self.p[p2])

```
